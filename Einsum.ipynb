```python
import torch
```

```python
x = torch.rand((2, 3))
x
```

```python
# permutation of tensors
torch.einsum('ij->ji', x)
```

```python
# Summantion
torch.einsum('ij->', x)
```

```python
# Column sum
torch.einsum('ij->j', x)
```

```python
# Row sum
torch.einsum('ij->i', x)
```

```python
# Matrix-vector multiplication
v = torch.rand((1, 3))

# (2, 3) MM (1, 3) -> (2, 3)
torch.einsum('ij,jk->ik', x, v)

# (2, 3) MM (3, 1) -> (2, 1)
torch.einsum('ij,kj->ik', x, v)
```

```python
# Matrix-vector multiplication
# x.mm(x.t())

torch.einsum("ij,kj->ik", x, x)
```

```python
# Dot product first row with first row of matrix
torch.einsum('i,i->', x[0], x[0])
```

```python
# Dot product with matrix
torch.einsum('ij,ij->', x, x)
```

```python
# Hadamard product (element-wise multiplication)
torch.einsum('ij,ij->ij', x, x)
```

```python
# Outer product
a = torch.rand((3))
b = torch.rand((5))
torch.einsum('i,j->ij', a, b)
```

```python
# Batch matrix multiplication
a = torch.rand((3,2,5))
b = torch.rand((3,5,3))
torch.einsum('ijk,ikl->ijl', a, b).shape
```

```python
# Matrix diagonal
x = torch.rand((3,3))
print(x)
torch.einsum('ii->', x)
```

```python
# Matrix trace
torch.einsum('ii->i', x)
```
