# Tensor math and comparision

```python
x = torch.tensor([1,2,3])
y = torch.tensor([9,8,7])
```

```python
z1 = torch.empty(3)
torch.add(x,y, out=z1)
z1
```

```python
z2 = torch.add(x, y)
z2
```

```python
z = x+y
z
```

```python
z = x-y
z
```

```python
z = torch.true_divide(x, y)
z
```

```python
t = torch.zeros(3)
t.add_(x)
```

```python
x
```

```python
# Exponentiation
x.pow(2), x**2
```

```python
# Simple comparison
x>0, x<0
```

```python
# Matrix multiplication
x1 = torch.rand((2,5))
x2 = torch.rand((5,3))
x3 = torch.mm(x1, x2)
x3
```

```python
x3 == x1.mm(x2)
```

```python
x3 == x1 @ x2
```

```python
# Matrix exponentiation
matrix_exp = torch.diag(torch.arange(1, 4))
matrix_exp.matrix_power(3)
```

```python
# Element wise multiplication
print(x, y)
z = x * y
print(z)
```

```python
# Dot product
z = torch.dot(x, y)
z
```

```python
# Batch matrix multiplication
batch = 32
n = 10
m = 20
p = 30

t1 = torch.rand((batch, n, m))
t2 = torch.rand((batch, m, p))
out_bmm = torch.bmm(t1, t2) # (batch, n, p)
```

```python
out_bmm.shape
```

```python
# Example of Broadcasting
x1 = torch.rand((5,5))
x2 = torch.rand((1,5))

x1-x2
```

```python
x1 ** x2
```
